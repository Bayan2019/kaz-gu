```python
import pandas as pd
from scipy.stats import chi2, chisquare
import numpy as np
```

# $\chi^2$ test

A $\chi^2$-squared test is a statistical hypothesis test used to determine if there is a significant difference between observed frequencies and expected frequencies in categorical data. 
It is used to test for the independence of two categorical variables (test of independence) or to see if a sample's frequency distribution matches a hypothesized distribution (goodness of fit test). 
If the resulting $p$-value is below the significance level (e.g., 0.05), the null hypothesis is rejected, suggesting the observed data is not as expected. 

```python
# Calculate PPF (quantile function)
# This finds the value x for which the CDF is equal to a given probability
ppf_value = chi2.ppf(0.05, df=1)
print(f"95th percentile: {ppf_value}")
```

```python
# Observed frequencies
observed = np.array([41, 34, 54, 39, 49, 45, 41, 33, 37, 41, 47, 39])

# Expected frequencies (e.g., for a fair six-sided die rolled 120 times)
expected = np.array([41.66, 41.66, 41.66, 41.66, 41.66, 41.66, 41.66, 41.66, 41.66, 41.66, 41.66, 41.66])

# Calculate the chi-square statistic and p-value
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")
```
