```python
import numpy as np
import pandas as pd
```

```python
# from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
from IPython.display import display, Latex
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([6, 16, 35, 55, 86])
df = pd.DataFrame({'x': x, 'y': y})
```

### Linear Estimation

```python
# 2. Add a constant to the independent variable for the intercept
# This creates the design matrix for statsmodels
X_with_intercept = sm.add_constant(x)
# 3. Create and fit the OLS (Ordinary Least Squares) model
model = sm.OLS(y, X_with_intercept)
results = model.fit()
# 4. Print the detailed summary
print(results.summary())
```

### Non-linear Estimation

```python
formula = 'y ~ x + I(x**2)'
```

```python
model = smf.ols(formula=formula, data=df)
results = model.fit()
```

```python
print(results.summary())
```

```python
predictions = results.predict(df)
predictions
```
